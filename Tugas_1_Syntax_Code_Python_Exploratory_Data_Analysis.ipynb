BARTPLOT HORIZONTAL BERDASARKAN TAHUN 1980 SAMPAI 2018

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv")
df = df[df["Year"].between(1980, 2018)]
x = df["Year"]
y = df["Total economic damage from natural disasters (US$)"]

# Barplot horizontal
plt.barh(x, y, color="blue", edgecolor="black")
plt.xlabel("Total economic damage from natural disasters (US$)")
plt.ylabel("Year")
plt.show()
```

BARTPLOT VERTIKAL BERDASARKAN TAHUN 1980 SAMPAI 2018

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv")
df = df[df["Year"].between(1980, 2018)]
x = df["Year"]
y = df["Total economic damage from natural disasters (US$)"]

# Barplot vertikal
plt.bar(x, y, color="red", edgecolor="black")
plt.xlabel("Tahun")
plt.ylabel("Total kerusakan ekonomi akibat bencana alam (USD)")
plt.show()
```

DISPLOT BERDASARKAN DATA DARI VARIABEL "TOTAL ECONOMIC DAMAGE FROM NATURAL DISARTER (US$)

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv")
x = df["Total economic damage from natural disasters (US$)"]
plt.hist(x, bins=100, color="green", edgecolor="black")
plt.xlabel("Total economic damage from natural disasters (US$)")
plt.ylabel("Frekuensi")
plt.show()
```

BOXPLOT

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv")
x = df["Total economic damage from natural disasters (US$)"]

# Boxplot
plt.boxplot(x)
plt.xlabel("Total economic damage from natural disasters (US$)")
plt.ylabel("Nilai")
plt.show()
```

SCATTERPLOT MENGGUNAKAN DATA NUMERIK TAHUN 1980 SAMPAI 2018

```python
df = df[df['Year'].between(1980, 2018)]
x = df["Total economic damage from natural disasters (US$)"]
y = df["Year"]

# Scatterplot
plt.scatter(x, y, color="orange", edgecolor="black")
plt.xlabel("Total economic damage from natural disasters (US$)")
plt.ylabel("Year")
plt.title("Scatterplot Untuk Total Economic Damage From Natural Disasters vs Year (1980-2018)")
plt.show()
```

PIE CHART UNTUK VARIABEL ENTITY

```python
x = df["Entity"]

#Pie Chart
plt.pie(df["Entity"].value_counts(), labels=df["Entity"].unique(), colors=["red", "green", "blue", "yellow", "purple"], autopct="%1.1f%%")
plt.title("Proporsi Jumlah Bencana Alam Berdasarkan Entitas")
plt.show()
```
